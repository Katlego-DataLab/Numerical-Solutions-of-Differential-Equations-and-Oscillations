```python
import numpy as np
import matplotlib.pyplot as plt


## Parameters

m = 1.0        # mass (kg)
k = 1.0        # spring constant (N/m)
dt = 0.01      # time step
t = np.arange(0, 20, dt)

# Initial conditions
x0 = 1.0       # initial displacement (m)
v0 = 0.0       # initial velocity (m/s)


## Euler Method

x_euler = np.zeros(len(t))
v_euler = np.zeros(len(t))
x_euler[0] = x0
v_euler[0] = v0

for i in range(1, len(t)):
    x_euler[i] = x_euler[i-1] + dt * v_euler[i-1]
    v_euler[i] = v_euler[i-1] - dt * (k/m) * x_euler[i-1]


## Runge-Kutta 2nd Order (RK2)

x_rk2 = np.zeros(len(t))
v_rk2 = np.zeros(len(t))
x_rk2[0] = x0
v_rk2[0] = v0

for i in range(1, len(t)):
    k1x = v_rk2[i-1]
    k1v = -(k/m) * x_rk2[i-1]

    k2x = v_rk2[i-1] + dt * k1v
    k2v = -(k/m) * (x_rk2[i-1] + dt * k1x)

    x_rk2[i] = x_rk2[i-1] + (dt / 2) * (k1x + k2x)
    v_rk2[i] = v_rk2[i-1] + (dt / 2) * (k1v + k2v)


## Figure 1: RK2 Method

plt.figure()
plt.plot(t, x_rk2, label='RK2 Method')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('Spring-Mass System using Runge-Kutta 2nd Order')
plt.legend()
plt.grid()
plt.show()


## Figure 2: Euler Method

plt.figure()
plt.plot(t, x_euler, label="Euler Method")
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title("Spring-Mass System using Euler's Method")
plt.legend()
plt.grid()
plt.show()


## Figure 3: Euler vs RK2

plt.figure()
plt.plot(t, x_euler, '--', label='Euler Method')
plt.plot(t, x_rk2, label='RK2 Method')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('Spring-Mass System: Euler vs RK2')
plt.legend()
plt.grid()
plt.show()
```
